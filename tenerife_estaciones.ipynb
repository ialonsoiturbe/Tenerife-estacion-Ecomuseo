Para hacer un análisis de los datos vamos a visualizar las medidas por fecha. Para ello primero vamos a dividir el df para cada sensor y comparar estos datos en un gráfico. Dividimos el dataset en varios subdataset porque si no no podemos realmente ver el valor de cada uno distinto y se pueden mezclar datos, separar los dataset nos permite analizar de forma clara y diferenciada cada tipo de medida.

```python
#Funcion para comparar los datos de cada sensor por estacion
def crear_comparacion(lista_estaciones, sensor):
    #Nos quedamos con la metrica que más nos interesa de cada sensor
    metricas_sensor = {
        "TEMP": "media",
        "HUM": "media",
        "WSP": "media",
        "RAIN": "total",
        "RAD": "total"      # o "maximo" si prefieres comparar el pico diario
    }

    metrica = metricas_sensor[sensor]

    series = []
    
    #dividimos por estaciones para que nos grafique todas de golpe
    for df in lista_estaciones:

        nombre = df["estacion_nombre"].iat[0]

        tabla = (
            df[df["sensor_alias"] == sensor]
            .pivot_table(
                index="fecha_observacion",
                columns="metrica",
                values="valor",
                aggfunc="first"
            )
        )

        if metrica in tabla.columns:
            series.append(tabla[metrica].rename(nombre))

    return pd.concat(series, axis=1)

#Agrupamos los sensores
temp = crear_comparacion(lista_estaciones, "TEMP")
hum  = crear_comparacion(lista_estaciones, "HUM")
wsp  = crear_comparacion(lista_estaciones, "WSP")
rain = crear_comparacion(lista_estaciones, "RAIN")
rad  = crear_comparacion(lista_estaciones, "RAD")
```

# SENSORES
Ahora que tenemos todos los datos separados por sensor y por estación, vamos a representarlos y enfrentarlos por los valores de las métricas.

```python
for nombre, df in {
    "Temperatura": temp,
    "Humedad": hum,
    "Viento": wsp,
    "Lluvia": rain,
    "Radiación": rad
}.items():

    plt.figure(figsize=(14,6))

    for estacion in df.columns:
        plt.plot(df.index, df[estacion], label=estacion)

    plt.title(nombre)
    plt.xlabel("Fecha")
    plt.ylabel(nombre)
    plt.grid(True)

    # Si tienes muchas estaciones
    plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left", fontsize=8)

    plt.tight_layout()
    plt.show()
```
